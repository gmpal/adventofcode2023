```python
import numpy as np
```

```python
with open('small.txt', 'r') as f:
    lines = f.readlines()
rows = len(lines)
cols = len(lines[0])
```

```python
grid = np.zeros((rows, cols), dtype=object)
```

```python
for line_idx, line in enumerate(lines):
    for element_idx, element in enumerate(line):
       grid[line_idx, element_idx] = element  
```

```python
def get_up(x, y):
    if x == 0:
        return None
    return grid[x - 1, y], x - 1, y

def get_down(x, y):
    if x == rows - 1:
        return None
    return grid[x + 1, y], x + 1, y

def get_left(x, y):
    if y == 0:
        return None
    return grid[x, y - 1], x, y - 1

def get_right(x, y):
    if y == cols - 1:
        return None
    return grid[x, y + 1], x, y + 1
```

```python
def find_s():
    for i in range(rows):
        for j in range(cols):
            if grid[i, j] == "S":
                return i, j
```

```python
def flow(x, y, direction, counter, pipes):
    counter += 1
    if direction == "up":
        next_, next_x, next_y  = get_up(x, y)
        if next_ == '|' :
            pipes.append((next_x, next_y))
            return (next_, next_x, next_y, "up", counter, pipes)
        elif next_ == 'F':
            pipes.append((next_x, next_y))
            return (next_, next_x, next_y, "right", counter, pipes)
        elif next_ == '7':
            pipes.append((next_x, next_y))
            return (next_, next_x, next_y, "left", counter, pipes)
        elif next_ == 'S':
            
            return (None, None, None, 'ok', counter, pipes)
        else:
            return (None, None, None, 'stop', counter, pipes)
    elif direction == "down":
        next_, next_x, next_y  = get_down(x, y)
        if next_ == '|' :
            pipes.append((next_x, next_y))
            return (next_, next_x, next_y, "down", counter, pipes)
        elif next_ == 'L':
            pipes.append((next_x, next_y))
            return (next_, next_x, next_y, "right", counter, pipes)
        elif next_ == 'J':
            pipes.append((next_x, next_y))
            return (next_, next_x, next_y, "left", counter, pipes)
        elif next_ == 'S':
            return (None, None, None, 'ok', counter, pipes)
        else:
            return (None, None, None, 'stop', counter, pipes)
    elif direction == "left":
        
        next_, next_x, next_y  = get_left(x, y)
        if next_ == '-' :
            pipes.append((next_x, next_y))
            return (next_, next_x, next_y, "left", counter, pipes)
        elif next_ == 'L':
            pipes.append((next_x, next_y))
            return (next_, next_x, next_y, "up", counter, pipes)
        elif next_ == 'F':
            pipes.append((next_x, next_y))
            return (next_, next_x, next_y, "down", counter, pipes)
        elif next_ == 'S':
            return (None, None, None, 'ok', counter, pipes)
        else:
            return (None, None, None, 'stop', counter, pipes)
    elif direction == "right":
        next_, next_x, next_y  = get_right(x, y)
        if next_ == '-' :
            pipes.append((next_x, next_y))
            return (next_, next_x, next_y, "right", counter, pipes)
        elif next_ == '7':
            pipes.append((next_x, next_y))
            return (next_, next_x, next_y, "down", counter, pipes)
        elif next_ == 'J':
            pipes.append((next_x, next_y))
            return (next_, next_x, next_y, "up", counter, pipes)
        elif next_ == 'S':
            return (None, None, None, 'ok', counter, pipes)
        else:
            return (None, None, None, 'stop', counter, pipes)
    else:
        raise ValueError("Invalid direction")

    
```
