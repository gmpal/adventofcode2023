```python
with open('input.txt') as f:
    lines = f.readlines()

#strip
lines = [x.strip() for x in lines]
```

```python
lines
```

```python
def get_up(x, y):
    return x - 1, y

def get_down(x, y):
    return x + 1, y

def get_left(x, y):
    return x, y - 1

def get_right(x, y):
    return x, y + 1
```

```python
def dig(x, y, direction, meters):
    path = []
    meters = int(meters)
    while meters != 0:
        if direction == 'U':
            x, y = get_up(x, y)    
        elif direction == 'D':
            x, y = get_down(x, y)
        elif direction == 'L':
            x, y = get_left(x, y)
        elif direction == 'R':
            x, y = get_right(x, y)
        path.append((x, y))
        meters -= 1
    return path
```

```python
full_path = []
for line in lines:
    direction = line.split(' ')[0]
    meters = line.split(' ')[1]
    print(meters)
    if len(full_path) == 0:
        x = 0
        y = 0
    else:
        x, y = full_path[-1]

    subpath = dig(x,y, direction, meters)
    full_path.extend(subpath)
```

```python
def is_point_inside_polygon(x, y, polygon):
    n = len(polygon)
    inside = False

    p1x, p1y = polygon[0]
    for i in range(n + 1):
        p2x, p2y = polygon[i % n]
        if y > min(p1y, p2y):
            if y <= max(p1y, p2y):
                if x <= max(p1x, p2x):
                    if p1y != p2y:
                        xinters = (y - p1y) * (p2x - p1x) / (p2y - p1y) + p1x
                    if p1x == p2x or x <= xinters:
                        inside = not inside
        p1x, p1y = p2x, p2y

    return inside

# Example usage:
# polygon = [(x1, y1), (x2, y2), ..., (xn, yn)]
# point = (a, b)
# is_inside = is_point_inside_polygon(point[0], point[1], polygon)
```

```python
from tqdm import tqdm

sum_ = 0
for x in tqdm(range(0, max_x)):
    for y in range(0, max_y):
        if grid[x, y] == '.':
            if is_point_inside_polygon(x, y, full_path):
                sum_ += 1
```

```python
full_path = []
for line in tqdm(lines):
    direction = line.split(' ')[0]
    hexa = line.split(' ')[2]
    actual_meters = int(hexa[2:-2], 16)
    actual_direction_int = hexa[-2]
    if actual_direction_int == '0':
        actual_direction = 'R'
    elif actual_direction_int == '1':
        actual_direction = 'D'
    elif actual_direction_int == '2':
        actual_direction = 'L'
    elif actual_direction_int == '3':
        actual_direction = 'U'
    if len(full_path) == 0:
        x = 0
        y = 0
    else:
        x, y = full_path[-1]

    subpath = dig(x,y, actual_direction, actual_meters)
    full_path.extend(subpath)
```
