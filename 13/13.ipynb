```python
import numpy as np
```

```python
with open('input.txt','r') as f:
    lines = f.readlines()
```

```python
counter = 0
games = {} 
games[counter] = []
for line in lines:
    if line == '\n':
        counter += 1
        games[counter] = []
    else:
        games[counter].append(line.strip())   
```

```python
grids = {}
for game_idx, game in games.items():
    rows = len(game)
    cols = len(game[0])
    grid = np.zeros((rows, cols), dtype=object)
    for line_idx, line in enumerate(game):
        for element_idx, element in enumerate(line):
            grid[line_idx, element_idx] = element  
    grids[game_idx] = grid
```

```python
def get_column(grid, col_idx):
    return grid[:, col_idx]
def get_row(grid, row_idx):
    return grid[row_idx, :]
```

```python
def get_consecutive_columns_pairs(grid):
    consecutive_columns_pairs = []
    for i in range(grid.shape[1] - 1):
        if (get_column(grid, i) == get_column(grid, i+1)).all(): #if two consecutive columns are the same
            consecutive_columns_pairs.append((i, i+1))
    return consecutive_columns_pairs

def get_consecutive_rows_pairs(grid):
    consecutive_rows_pairs = []
    for i in range(grid.shape[0] - 1):
        if (get_row(grid, i) == get_row(grid, i+1)).all(): #if two consecutive rows are the same
            consecutive_rows_pairs.append((i, i+1))
    return consecutive_rows_pairs
```

```python
def check_palyndrome_cols(grid, pair):
    first_col_idx,second_col_idx = pair[0], pair[1]
    while first_col_idx >= 0 and second_col_idx < grid.shape[1]:
        if (get_column(grid, first_col_idx) == get_column(grid, second_col_idx)).all():
            first_col_idx -= 1
            second_col_idx += 1
        else:
            return False
    return True

def check_palyndrome_rows(grid, pair):
    first_row_idx,second_row_idx = pair[0], pair[1]
    while first_row_idx >= 0 and second_row_idx < grid.shape[0]:
        if (get_row(grid, first_row_idx) == get_row(grid, second_row_idx)).all():
            first_row_idx -= 1
            second_row_idx += 1
        else:
            return False
    return True
```

```python
to_sum = {}
simmetries_rows = {}
simmetries_cols = {}

for grid_idx, grid in grids.items():
    cons_cols = get_consecutive_columns_pairs(grid)
    cons_rows = get_consecutive_rows_pairs(grid)

    mirror_columns_left = 0
    for pair in cons_cols:
        if check_palyndrome_cols(grid, pair):
            simmetries_cols[grid_idx] = pair
            mirror_columns_left = pair[0] + 1 
    
    mirror_columns_up = 0
    for pair in cons_rows:
        if check_palyndrome_rows(grid, pair):
            simmetries_rows[grid_idx] = pair
            mirror_columns_up = pair[1]       

    to_sum[grid_idx] = (mirror_columns_left) + 100 * (mirror_columns_up)

sum_ = 0
for _, value in to_sum.items():
    sum_ += value
print(sum_)
```
